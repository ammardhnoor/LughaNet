```python
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer

def translate(text, model_name="facebook/m2m100_418M"):
    # Initialize the tokenizer and model
    tokenizer = M2M100Tokenizer.from_pretrained(model_name)
    model = M2M100ForConditionalGeneration.from_pretrained(model_name)

    # Set the source and target languages
    tokenizer.src_lang = "en"
    model.config.forced_bos_token_id = tokenizer.get_lang_id("ar")

    # Encode the text
    encoded = tokenizer(text, return_tensors="pt")

    # Generate translation
    translated_tokens = model.generate(**encoded)

    # Decode and return the translated text
    translated_text = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)
    return translated_text

# Example usage
english_word = "peace"
arabic_translation = translate(english_word)
print(arabic_translation)
```

```python
import json
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer

def translate(text, model_name="facebook/m2m100_418M"):
    # Initialize the tokenizer and model
    tokenizer = M2M100Tokenizer.from_pretrained(model_name)
    model = M2M100ForConditionalGeneration.from_pretrained(model_name)

    # Set the source and target languages
    tokenizer.src_lang = "en"
    model.config.forced_bos_token_id = tokenizer.get_lang_id("ar")

    # Encode the text
    encoded = tokenizer(text, return_tensors="pt")

    # Generate translation
    translated_tokens = model.generate(**encoded)

    # Decode and return the translated text
    translated_text = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)
    return translated_text

# Load English words from JSON file
with open("E:/تحدددددددددددديث task1/split not cover/part2.json", 'r', encoding='utf-8') as file:
    english_words = json.load(file)

# Select the first 10 words
first_10_words = english_words[:2410]

# Translate each word and create a dictionary mapping English to Arabic
translation_mapping = {word.strip(): translate(word.strip()) for word in first_10_words}

# Display each translation
for english, arabic in translation_mapping.items():
    print(f"{english} : {arabic}")

# Optionally, save the translation mapping to a new JSON file
with open("E:/تحدددددددددددديث task1/largemode translein/m2m/m222m22.json", 'w', encoding='utf-8') as file:
    json.dump(translation_mapping, file, ensure_ascii=False)

print("Translation of the  words completed.")
```
